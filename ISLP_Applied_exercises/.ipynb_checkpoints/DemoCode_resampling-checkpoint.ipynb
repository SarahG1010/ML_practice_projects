### Confusion matrix construction:
The confusion_matrix() function in scikit-learn computes a confusion matrix to evaluate the accuracy of a classification.

Here's the basic syntax:

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_true, y_pred, labels=None, sample_weight=None, normalize=None)
```

* **y_true**: This is the ground truth (correct) target values.
* **y_pred**: This is the predicted targets as returned by a classifier.
* **labels**: List of labels to index the matrix. This may be used to reorder or select a subset of labels. If None is given, those that appear at least once in y_true or y_pred are used in sorted order.
* **sample_weight**: Sample weights.
* **normalize**: Normalization can be applied by setting this parameter to 'true' or 'all'.

The returned conf_matrix is a 2D array where the element in the i-th row and j-th column represents the number of samples with true label being the i-th class and predicted label being the j-th class.


```python
from sklearn.metrics import confusion_matrix

# Example true labels and predicted labels
true_labels = ['cat', 'dog', 'cat', 'bird', 'dog', 'bird']
predicted_labels = ['cat', 'dog', 'cat', 'bird', 'dog', 'dog']

# Construct confusion matrix
conf_matrix = confusion_matrix(true_labels, predicted_labels)

# Print confusion matrix
print("Confusion matrix:")
print(conf_matrix)
```

### OUTPUT
Confusion matrix:

[[2 0 0]

 [0 2 0]
 
 [1 0 1]]

In this matrix, rows represent the true labels and columns represent the predicted labels. For example, the value at row 1, column 1 (2) represents the number of samples where the true label is 'cat' and the predicted label is also 'cat'.

#### why the value at row 1, column 1 (2) represents cat?
In the context of the confusion matrix generated by scikit-learn's confusion_matrix() function, the rows and columns represent the classes (labels) being predicted by your model.

In the example confusion matrix generated above, The rows and columns are indexed **based on the order of the unique classes encountered in your data**. In this case, the unique classes are ['cat', 'dog', 'bird'].

So, in the confusion matrix:


The value at row 1, column 1 (indexing is 0-based) represents the count of samples where the true label is 'cat' and the predicted label is also 'cat'.
The value at row 2, column 2 represents the count of samples where the true label is 'dog' and the predicted label is also 'dog'.
